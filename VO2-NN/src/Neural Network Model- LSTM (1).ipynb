```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold,LeaveOneOut,cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.layers import Dense, LSTM
from keras.optimizers import Adam
from keras.models import Sequential
from sklearn.metrics import r2_score, mean_squared_error
from sklearn import linear_model, svm
from numpy import mean, std, absolute
from keras.models import Sequential
#Note: Use cross validation
#, split data different test and train Random_state=None
#(validation - small dataset)) Done
```

```python
seq= pd.read_csv('Data.csv').values
seq = np.array(seq)
seq
```

```python
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column (Column Gender) 1 - Males, 0 - Females
seq[:,3] = np.array(labelencoder.fit_transform(seq[:,3]))
seq
```

```python
X, y = seq[:, 0:6], seq[:, 6]
X = np.asarray(X).astype(np.float32)
X = X.reshape((len(X), 6, 1)) 
y = np.asarray(y).astype(np.float32)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=1) # None, 0, 1, 2
# splits data randomly everytime 
print(X_train)
print("test:",X_test)
```

```python
# define model
model = Sequential()
model.add(LSTM(32, input_shape=(6,1), activation='relu',return_sequences=True))#return_sequences=True
model.add(LSTM(16,activation='relu')) #uncomment these for model from Zignoli et al 2020
#model.add(LSTM(32))
model.add(Dense(1, activation='linear'))
model.compile(optimizer = 'adam', loss = 'mean_squared_logarithmic_error')
model.fit(X, y, epochs = 5000, batch_size=17,validation_data=(X_test, y_test))
```

```python
# make predictions
yhat = model.predict(X_test)
print(yhat)
```

```python
#Visualising the Actual and predicted Result
plt.plot(y_test, 'o', color = 'green', label = 'Actual')
plt.plot(yhat,'o', color = 'blue', label = 'Predicted')
plt.xlabel('Sample Number ')
plt.ylabel('VO2max')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```

```python
print("mean squared error: ", mean_squared_error(y_test, yhat)) #compute MSE
print("r_square score: ", r2_score(y_test, yhat)) # compute R2 score (accuracy for regression tasks or continuous data)
```

```python
train_error =  np.abs(y_test - yhat)
mean_error = np.mean(train_error)
min_error = np.min(train_error)
max_error = np.max(train_error)
std_error = np.std(train_error)
print("std_error: ",std_error)
print("mean_error: ",mean_error)
print("min_error: ",min_error)
print("max_error: ",max_error)
```
